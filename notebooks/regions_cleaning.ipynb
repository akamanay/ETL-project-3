## ETL operation for regions

```python
import pandas as pd

# Loaded variable 'df' from URI
URI = '../datas/regions.xlsx'
df = pd.read_excel(URI)

# Remove the prefix 'ERROR' in regions_name
df['region_name'] = df['region_name'].str.replace('^ERROR','', regex=True)

# Capitalize the first letter of region_name since it is a region name
df['region_name'] = df['region_name'].str.capitalize()
df
```

## Generate the cleaned file

```python
df.to_excel('../output_datas/regions_v2.xlsx',index=False)
```
