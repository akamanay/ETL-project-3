# Fill blanks with null

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../datas/supply_chain.csv")

df.dropna(inplace=True)

df['supplier_id'] = np.array(df['supplier_id'], dtype='i')
df['delivery_date'] = pd.to_datetime(df['delivery_date'], errors='coerce')

df_clean = df
df_clean
```

# Save cleaned data

```python
path = '../output_datas/supply_chain_v2.xlsx'
df_clean.to_excel(path, index=False)
```
